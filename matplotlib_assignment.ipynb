```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x,y)

plt.title("Simple Line Plot")
plt.xlabel("X-axis (Independent Variable)")
plt.ylabel("Y-axis (Dependent Variable)")
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, color=['blue', 'yellow', 'green', 'red'])

plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.ylim(0, 100)  # Set Y-axis limit for better scale

plt.show()
```

```python
import matplotlib.pyplot as plt

regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

plt.pie(
    revenue, 
    labels=regions
)

# Step 3: Add a title
plt.title("Revenue Distribution by Region")


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randint(1, 101, size=1000)

plt.hist(data, bins=10, color='skyblue', edgecolor='black')

plt.title("Frequency Distribution of Random Integers (1–100)")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.grid(True)


plt.show()
```
